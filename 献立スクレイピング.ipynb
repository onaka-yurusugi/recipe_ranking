```
import requests
from bs4 import BeautifulSoup

# ベースとなる楽天レシピのURL
base_url = 'https://recipe.rakuten.co.jp/search/'

# 献立検索関数を作る
def Cook_Search(word):
    url = base_url + word
    res = requests.get(url)

    # レスポンスの HTML から BeautifulSoup オブジェクトを作る
    soup = BeautifulSoup(res.text, 'html.parser')

    # レシピのタイトル名を取得
    elems = soup.select(".recipe_ranking__recipe_title")

    # レシピのリンクを取得
    links = [url.get('href') for url in soup.select('.recipe_ranking__link')]

    rank = 0
    for elem in elems:
        rank += 1
        print(str(rank) + "位！")
        print(elem.getText())
        print("https://recipe.rakuten.co.jp" + links[rank - 1])
        print("-" * 80)

# 食材はスペース区切りで入力するように
# サンプル
Cook_Search("じゃがいも")
```
